```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_marketing = pd.read_csv('marketing_customer_analysis-round2.csv')
df_marketing
```

```python
#Show DataFrame info
df_marketing.info()
```

```python
#Describe DataFrame
df_marketing.describe(include="all").T
```

```python
#Show a plot of the total number of responses
sns.displot(df_marketing['response'],bins=2)
plt.show()
```

```python
def function (x):
    if x == 'Yes':
        return 1
    else:
        return 0

df_marketing['response rate'] = df_marketing['response'].apply(function)
df_marketing
```

```python
sns.barplot(x='sales channel', y='response rate', data = df_marketing)
plt.show()
```

```python
sns.countplot(x='sales channel', hue='response', data = df_marketing)
plt.show()
```

```python
sns.barplot(x='response', y='total claim amount', data = df_marketing)
```

```python
sns.catplot(x="response",y="total claim amount", data=df_marketing, jitter="0.4")
plt.show()
```

```python
sns.barplot(x='response', y='income', data = df_marketing)
plt.show()
```

```python
sns.catplot(x="response",y="income", data=df_marketing, jitter="0.4")
plt.show()
```
